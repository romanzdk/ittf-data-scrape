```
import requests
import pandas as pd
import bs4
import csv
```

```
url = "https://ranking.ittf.com/public/s/ranking/csv?category=SEN&typeGender=W%3BSINGLES&year="
```

```
df = pd.DataFrame()
```

```
for i in range(2001, 2020):
  for j in range(1, 13):
    url2 = url + str(i) + "&month=" + str(j)
    df = df.append(pd.read_csv(url2))
```

```
df
```

```
df2 = pd.DataFrame()
```

```
for i in range(2019, 2021):
  for j in range(1, 50):
    url2 = url + str(i) + "&week=" + str(j)
    df2 = df2.append(pd.read_csv(url2))
```

```
complete = df.append(df2)
```

```
complete
```

```
complete.to_csv('ittf_rankings_women.csv', index=False)
```

```
df = pd.DataFrame()
```

```
for i in range(0, 37620, 10):
  url = r'https://results.ittf.link/index.php?option=com_fabrik&view=list&listid=35&Itemid=155&resetfilters=0&clearordering=0&clearfilters=0&limitstart35='
  url = url + str(i)
  tables = pd.read_html(url)
  table = tables[1]
  df = df.append(table.iloc[:10, :9])
```
